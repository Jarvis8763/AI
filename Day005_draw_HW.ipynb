請根據 Lena 圖做以下處理

- 對明亮度做直方圖均衡處理
- 水平鏡像 + 縮放處理 (0.5 倍)
- 畫出人物矩形邊框

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('lena.png')
```

```python
# 對明亮度做直方圖均衡處理
img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img[...,2]=cv2.equalizeHist(img[...,2])
img=cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
```

```python
# 畫出人物矩形邊框
cv2.rectangle(
img, # 圖片
(60, 40), # 左上角的座標
(420, 510), # 右下角的座標
(255, 255, 0), # RGB 顏色
3) # 線的粗細程度

# 水平鏡像 + 縮放處理 (0.5 倍)
img=img[:,::-1,:]
plt.imshow(cv2.resize(img, None,fx=0.5, fy=0.5))
```
